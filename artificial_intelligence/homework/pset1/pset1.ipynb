## Question 2 Part 1

### Solution:
$\begin{aligned}
\hat\omega_1&=\arg\min_\omega\mathcal L_1(y,f(\omega;x_i))\\
&=\arg\min_\omega\frac{1}{2}(y-f(\omega;x))^2\\
&=\arg\min_\omega\frac{1}{2}(y-\phi(x)\cdot\omega)^2\\
\end{aligned}$  
$\begin{aligned}
\nabla\mathcal L_1(y,f(\omega;x)) &= \frac{d}{d\omega}\left(\frac{1}{2}(y-\phi(x)\cdot\omega)^2\right)\\
&= \frac{d}{d\omega}\left(\begin{array}{c}\frac{1}{2}(y_1-\phi(x_1)\cdot\omega)^2 \\ \frac{1}{2}(y_2-\phi(x_2)\cdot\omega)^2 \\ \cdots \\ \frac{1}{2}(y_d-\phi(x_d)\cdot\omega)^2\end{array}\right)\\
&= \left(\begin{array}{c}-\phi(x_1)(y_1-\phi(x_1)\cdot\omega) \\ -\phi(x_2)(y_2-\phi(x_2)\cdot\omega) \\ \cdots \\ -\phi(x_d)(y_d-\phi(x_d)\cdot\omega)\end{array}\right)\\
&=-\phi(x)(y-\phi(x)\cdot\omega)
\end{aligned}$
$\begin{aligned}
\nabla\mathcal L_n(y,f(\omega;x)) &= \frac{1}{N}\sum_{(x,y)\in S_N}-\phi(x)(y-\phi(x)\cdot\omega)\\
&= \frac{1}{N}\sum_{(x,y)\in S_N}-\phi(x)\cdot y+\phi(x)\cdot\phi(x)\cdot\omega\\
&=\frac{1}{N}(\Phi^T\Phi\ \omega-\Phi Y)
\end{aligned}$  
$\Phi$ as defined by the problem statement.  
When $\nabla\mathcal L_n(y,f(\omega;x))=0$,  
$\begin{aligned}
\frac{1}{N}(\Phi^T\Phi\ \hat\omega-\Phi Y) &= 0\\
\Phi^T\Phi\ \hat\omega-\Phi Y&= 0\\
\Phi^T\Phi\ \hat\omega &= \Phi Y \\
\hat\omega &= (\Phi^T\Phi)^{-1}\Phi Y
\end{aligned}$ 

## Question 2 Part 2

### Solution:

$\begin{aligned}
f(z)&=\phi(z)\cdot\omega\\
&= \phi(z)\cdot\Phi^Tv\\
&=\mathcal K(z,x)\cdot v
\end{aligned}$  
$K(x_i,x_j)= K(x_j,x_i)=\Phi\cdot\Phi^T$  
$K$ is symmetrical.  
Based on Question 2 Part 1,  
$\begin{aligned}
\hat v&=(K^TK)^{-1}KY\\
&=(KK)^{-1}KY\\
&=K^{-1}(K^{-1}K)Y\\
&=K^{-1}Y
\end{aligned}$  
$f(z)=\mathcal K(z,x)\cdot K^{-1}Y$

## Question 3 Part I

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from random import shuffle
from math import ceil
```

```python
def load_dat(dir):
    with open(dir,"r") as input:
        lines = input.readlines()
        mat = np.array([line.split() for line in lines], dtype=float).T
        x = mat[:-1].T
        y = mat[-1].T
    return x,y

x1, y1 = load_dat("dataLinReg1D.txt")
x2, y2 = load_dat("dataLinReg2D.txt")
```

```python
def append_bias(x,b=None):
    tmpx = x.T
    if not b:
        b = np.ones(tmpx.shape[1])
    return np.vstack((tmpx,b)).T

# print(append_bias(x1))
```

```python
def regression(x,y):

    newx = append_bias(x)
    theta1 = np.linalg.inv(np.matmul(newx.T,newx))
    theta2 = np.matmul(theta1,newx.T)
    theta3 = np.matmul(theta2,y)
    return theta3

print(regression(x1,y1))
print(regression(x2,y2))
```

```python
def plot2d(x,y):
    theta = regression(x,y)
    predy = np.dot(append_bias(x),theta)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.scatter(x,y, c="r")
    ax.scatter(x,predy,c="b")
    plt.show()
    
plot2d(x1,y1)
```

```python
def plot3d(x,y):
    theta = regression(x,y)
    predy = np.dot(append_bias(x),theta)
    fig = plt.figure()
    ax = fig.add_subplot(111, projection="3d")
    x1 = x.T[0]
    x2 = x.T[1]
    ax.scatter(x1,x2,y,c="r")
    ax.scatter(x1,x2,predy,c="b")
    plt.show()
    
plot3d(x2,y2)
```

## Question 3 Part II

### Solution

#### Without Noise

```python
def ridge_regression(x,y,lamb=0):
    newx = append_bias(x)
    theta1 = np.linalg.inv(np.matmul(newx.T,newx)+lamb*np.eye(newx.shape[1]))
    theta2 = np.matmul(theta1,newx.T)
    theta3 = np.matmul(theta2,y)
    return theta3

# print(ridge_regression(x1,y1))
# print(ridge_regression(x2,y2))

def mse(theta,y):
    return np.square((theta - y)).mean(axis=0)
    

def cross_val(x,y,n,fxn,lamb=0):
    indlist = [i for i in range(x.shape[0])]
    shuffle(indlist)
    trainerr = []
    validerr = []
    d =  ceil(x.shape[0]/n)
    for j in range(10):
        traintot = 0
        validtot = 0
        for i in range(0,x.shape[0],d):
            if i == 0:
                trainx = x[indlist[i:i+d]]
                trainy = y[indlist[i:i+d]]
                validx = x[indlist[i+d:]]
                validy = y[indlist[i+d:]]
            elif (i+d) < x.shape[0]-1:
                trainx = x[indlist[i:i+d]]
                trainy = y[indlist[i:i+d]]
                validx = np.concatenate((x[indlist[:i]], x[indlist[i+d:]]),axis=0)
                validy = np.concatenate((y[indlist[:i]], y[indlist[i+d:]]),axis=0)
            else:
                trainx = x[indlist[i:]]
                trainy = y[indlist[i:]]
                validx = x[indlist[:i]]
                validy = y[indlist[:i]]
            theta = fxn(trainx,trainy,lamb).T
            trainpredy = np.dot(append_bias(trainx),theta)
            validpredy = np.dot(append_bias(validx),theta)
            traintot += mse(trainpredy,trainy)
            validtot += mse(validpredy,validy)
        trainerr.append(traintot/n)
        validerr.append(validtot/n)
    return np.mean(trainerr),np.mean(validerr),np.std(trainerr),np.std(validerr)

def get_lamb(lamblist,meanlist):
    tmpi = [i for i in range(len(meanlist)) if meanlist[i] > 0]
    tmpx = [i for i in meanlist if i > 0]
    return lamblist[tmpi[tmpx.index(min(tmpx))]]
    

# print(cross_val(x1,y1,5,ridge_regression))
trainminlamb = []
validminlamb = []    
```

```python
lamblist = [(10-((10-0.0001)/1000)*i) for i in range(1000)[::-1]] # linear space
# lamblist = np.logspace(-4,1,1000) # logarithmic space

def find_waldo(x,y,n,fxn,lamblist):
    trainmean = []
    trainstd = []
    validmean = []
    validstd = []
    for lamb in lamblist:
        a,b,c,d = cross_val(x,y,n,fxn,lamb)
        trainmean.append(a)
        trainstd.append(b)
        validmean.append(c)
        validstd.append(d)
    return trainmean,trainstd,validmean,validstd

trainmean,trainstd,validmean,validstd = find_waldo(x2,y2,5,ridge_regression,lamblist)
```

##### Training Error (Polynomial Features)

```python
def plot(xrange,mean,std,scale=1,title=None):
    scale = ceil(len(xrange)*scale)
    xrange = xrange[:scale]
    mean = mean[:scale]
    std = std[:scale]
    fig = plt.figure()
    ax = fig.add_subplot(111)
    upper = [mean[i]+std[i] for i in range(len(mean))]
    lower = [mean[i]-std[i] for i in range(len(mean))]
    ax.plot(xrange,upper,c="b")
    ax.plot(xrange,lower,c="b")
    ax.bar(xrange,mean,zorder=10,color="orange")
    ax.set_xscale("log")
    ax.set_xlim(xmin=min(xrange))
    if title:
        plt.title(title)
    plt.show()

print("Min. Lambda: %s" % get_lamb(lamblist,trainmean))
plot(lamblist,trainmean,trainstd,title="Log Scale, Scale: 1")
plot(lamblist,trainmean,trainstd,scale=0.75,title="Log Scale, Scale: 0.75")
# plot(lamblist,trainmean,trainstd,scale=0.2,title="Log Scale, Scale: 0.2")
```

From most of the histograms, there is no visible difference as to the optimal value of $\lambda$. Most parameters show no change or increasing trend in MSE values.  

The only MSE histogram that showed any sign of convexity is the Training Error graph with RBF Features and no added noise. The optimal $\lambda$ found had the value of around 100. Even with noise, this trend can be seen with smaller intensity.  

All the Validation Errors were not of much help as the validation set size is 20 for each randomised 5 fold cross-validation cycle.  

For all error graphs with polynomial feature sets, the error tended to increase with respect to $\lambda$ when $\lambda$ is more than 0.  

With noise, all standard deviation of the MSE values increased significantly and its mean becomes relatively stable. This is most probably caused by the noise drowning out the actual values of $y$ too significantly when taking many cycles of 5 fold cross-validation.  

It is worth noting the RBF feature is more tolerant to the noise, but the noise is still too significant to find the optimal value of $\lambda$ reliably.
